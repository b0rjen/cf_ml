```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/calidad_agua.csv', encoding='ISO-8859-1', delimiter=';')
df.head()
```

```python
df.info()
```

```python
df.head(11)
```

```python
df = df.drop(columns=['CCAA', 'PROVINCIA'])
# voy a quitar estas columnas que no aportan
```

```python
df = df.dropna(axis=1, thresh=(df.shape[1] - 5))
# quito los que tengan más de seis nulos
```

```python
df = df.fillna(0)
# relleno el resto de ceros
```

```python
df.head(11)
```

```python
df['AÑO'] = pd.to_datetime(df['AÑO'], format='%Y')
df.set_index('AÑO')
# paso la columna año a index y la transformo en datetime
```

```python
df.dtypes
```

```python
# salen todo objects. se ve que el separador de decimales son comas, lo reemplazo por puntos y paso a número
df['PM_CALIDAD_AGUA_BUENA'] = df['PM_CALIDAD_AGUA_BUENA'].str.replace(',', '.').astype(float)
df['PM_CALIDAD_AGUA_EXCELENTE'] = df['PM_CALIDAD_AGUA_EXCELENTE'].str.replace(',', '.').astype(float)
df['PM_CALIDAD_AGUA_SIN_CALIFICAR'] = df['PM_CALIDAD_AGUA_SIN_CALIFICAR'].str.replace(',', '.').astype(float)
```

```python
# voy a graficar sólo los datos de calidad del agua, el resto no tiene interés

data_to_plot = df[['PM_CALIDAD_AGUA_BUENA', 'PM_CALIDAD_AGUA_EXCELENTE', 'PM_CALIDAD_AGUA_SIN_CALIFICAR']]

# Crear un gráfico de barras apiladas
data_to_plot.plot(kind='bar', stacked=True, figsize=(10, 7))

# Configurar los títulos y etiquetas
plt.title('Calidad del Agua por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)  # Rota las etiquetas del eje x para mejor legibilidad

# Mostrar el gráfico
plt.tight_layout()  # Ajusta automáticamente los parámetros del subplot para que el gráfico se ajuste al área disponible
plt.show()
```
