```python
import pandas as pd
```

```python
df = pd.read_excel('../data/turismo_receptor_mun_pais.xlsx')
```

```python
df.head()
```

```python

pais_turistas_new = df.groupby('PAIS_ORIGEN')['TURISTAS'].sum()

top_paises_new = pais_turistas_new.sort_values(ascending=False).head(5)

top_paises_new
```

```python
pais_turistas_new = df.groupby('PAIS_ORIGEN')['TURISTAS'].sum()
top_paises_new = pais_turistas_new.sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
# Calculamos el total de turistas y el total de los top 5 países para encontrar el "resto_de_paises"
total_turistas = pais_turistas_new.sum()
total_top_paises = top_paises_new.sum()
resto_de_paises = total_turistas - total_top_paises
```

```python
# Creamos una nueva Serie para "Resto de países"
resto_series = pd.Series({'Resto de países': resto_de_paises})

# Usamos pd.concat para unir la serie de top_paises_new con la serie de resto_de_paises
top_paises_con_resto = pd.concat([top_paises_new, resto_series])

# Ahora, top_paises_con_resto debería ser una Serie que incluye tanto los top países como el "resto de países"
```

```python
# Generamos el primer pie chart para los top 5 países
top_paises_new.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Países por Turistas')
plt.ylabel('')  # Eliminar la etiqueta del eje y
plt.show()
```

```python
# Generamos el segundo pie chart incluyendo el "resto_de_paises"
top_paises_con_resto.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Países por Turistas + Resto de Países')
plt.ylabel('')  # Eliminar la etiqueta del eje y
plt.show()
```
